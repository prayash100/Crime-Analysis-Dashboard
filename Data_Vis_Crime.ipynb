```python
import pandas as pd
df = pd.read_csv('Crime_Data.csv')
```

```python
df.drop(['Date Rptd'], axis=1, inplace=True)
```

```python
df.drop(['DATE OCC'], axis=1, inplace=True)
```

```python
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p')

# Extract only the year
df['R.Year'] = df['Date Rptd'].dt.year

# Extract only the month
df['R.Month'] = df['Date Rptd'].dt.month

df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')

# Extract only the year
df['O.Year'] = df['DATE OCC'].dt.year

# Extract only the month
df['O.Month'] = df['DATE OCC'].dt.month
```
